# **NLP PRACTICAL-7 WordCloud with Image**

### Aim:To write a program to implement word cloud of English Text.

Tag Cloud or WordCloud is a data visualization technique used for representing text data in which the size of each word indicates its frequency or importance. Significant textual data points can be highlighted using a word cloud. 

Collocations are phrases or expressions containing multiple words, that are highly likely to co-occur. For example — ‘social media’, ‘school holiday’, ‘machine learning’, ‘Universal Studios Singapore’, etc.

Bigrams: Bigram is 2 consecutive words in a sentence. E.g. “The boy is playing football”. The bigrams here are:


The boy
Boy is
Is playing
Playing football


Trigrams: Trigram is 3 consecutive words in a sentence. For the above example trigrams will be:

The boy is
Boy is playing
Is playing football

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as pPlot
from wordcloud import WordCloud, STOPWORDS
import numpy as npy
from PIL import Image
dataset = open("/content/movie list.txt", "r").read()
dataset = dataset.upper()
```

```python
maskArray = npy.array(Image.open("/content/video-300x300.jpg"))
cloud = WordCloud(background_color = "mediumpurple", min_font_size=5, colormap='copper',
max_font_size=70,font_path='/content/LEMONMILK-Medium.otf',max_words = 2000, width=1400, height=900,
collocations=True, mask = maskArray, stopwords = set(STOPWORDS)).generate(dataset)                
```

```python
# plot the WordCloud image                      
plt.figure(figsize = (6, 10), facecolor = None)
plt.imshow(cloud)
plt.axis("off")
```
